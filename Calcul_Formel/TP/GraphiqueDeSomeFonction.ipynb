```python
x = var('x')
P1(x) = x^5 - 2*x^4 + 3*x +1
P2(x) = x^4 - 2*x^3 +x +1
```

```python
print('2->1) Degre, des dominants de P = P1P2')
```

```python
P = P1*P2; show('p : ', P)
```

```python
d = P.degree(x)
```

```python
print('Le degree de P est, ', d)
```

```python
coef = P1.leading_coefficient(x)*P2.leading_coefficient(x)
```

```python
print('Le coeficiant dominant de P est, ', coef )
```
